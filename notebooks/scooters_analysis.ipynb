```python
from sqlalchemy import create_engine, text
import pandas as pd

# Assuming 'engine' is defined elsewhere in your script as an SQLAlchemy engine

count_squery = '''
SELECT companyname as company_name, COUNT(DISTINCT sumdid) as count_scooters
FROM scooters
WHERE sumdgroup <> 'bicycle'
GROUP BY companyname;
'''

with engine.connect() as connection:
    count_scoot = pd.read_sql(text(count_squery), con=connection)
    
time_squery = '''
SELECT companyname as company_name, COUNT(DISTINCT sumdid) AS count_scooters, date(pubdatetime) as date
FROM scooters
WHERE sumdgroup <> 'bicycle'
GROUP BY date, company_name
'''

with engine.connect() as connection:
    time_scoot = pd.read_sql(text(time_squery), con=connection)
    
print(count_scoot)
print(time_scoot.head())
```
